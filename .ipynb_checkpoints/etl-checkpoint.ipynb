```python
from pathlib import Path
import json
import requests
```

```python
PATH = "https://api.nal.usda.gov/fdc/v1/foods/search"
API_KEY = "6o7M2ay7ihJ6nIcVSUjsVpDc1hi2AqKJv0ix38Mg"
```

```python
def get_totalpage():
    url = f"{PATH}?api_key={API_KEY}&pageNumber=1&pageSize=100"
    data = requests.get(url).json()
    return data['totalPages']
```

```python
pages = get_totalpage()
foods_list = []

for page in range(1, 3):
    url = f"{PATH}?api_key={API_KEY}&pageNumber={page}&pageSize=100"
    data = requests.get(url).json()
    foods = data['foods']
    foods_list.append(foods)
```
